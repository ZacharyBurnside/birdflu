```python
import pandas as pd

# assuming you already have df loaded

# extract the first number before 'sold' or 'times' and convert to int
df['sold_count'] = (
    df['message']
      .str.extract(r'(\d+)\s*(?:sold|times)', expand=False)
      .astype('Int64')
)


# quick check
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ensure your date column is datetime (replace 'collection_date' if different)
df['collection_date'] = pd.to_datetime(df['collection_date'])

# sum sold_count per day
daily = df.groupby('collection_date')['sold_count'].sum()

# plot
daily.plot(figsize=(10,5), marker='o')
plt.xlabel('Date')
plt.ylabel('Sold Count')
plt.title('Sold Count Over Time')
plt.tight_layout()
plt.show()
```
